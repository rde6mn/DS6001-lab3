```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
cocktail_url = 'http://www.thecocktaildb.com/api/json/v1/1/filter.php?c=Cocktail'
```

```python
r=requests.get(cocktail_url)
r
```

```python
r.text
```

```python
myjson = json.loads(r.text)
#myjson = r.json()
myjson
```

```python
myjson["drinks"][0]["strDrink"]
```

```python
cocktails_df = pd.json_normalize(myjson, record_path=['drinks'])
```

```python
cocktails_df
```

## User Agent

```python
useragent_url = 'https://httpbin.org/user-agent'
```

```python
r=requests.get(useragent_url)
r
```

```python
useragent = r.json()['user-agent']
useragent
```

```python
headers = {'User-Agent': useragent,
           'From': 'rde6mn@virginia.edu'}
headers['User-Agent'] = 'rde6mn/0.0 (rde6mn@virginia.edu) python-requests/2.32.4'
```

## Wikipedia

```python
wikiurl = 'https://en.wikipedia.org/w/api.php?action=query&list=search&srsearch=Virginia&format=json&srlimit=500'
r = requests.get(wikiurl, headers=headers)
r
```

```python
r.text
```

```python
wiki_df = pd.json_normalize(r.json(), record_path=['query','search'])
wiki_df
```
